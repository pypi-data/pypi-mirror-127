```python
%load_ext autoreload
%autoreload 2

import time
import pysbrl
import numpy as np
from imodels import BayesianRuleListClassifier
from imodels.util.discretization.simple import SimpleDiscretizer
from imodels.util.rule import get_feature_dict
from experiments.util import get_clean_dataset
```

```python
X, y, fnames = get_clean_dataset("../data/credit_g.csv")
```

```python
feature_dict_ = get_feature_dict(X.shape[1], fnames)
feature_placeholders = np.array(list(feature_dict_.keys()))
feature_names = np.array(list(feature_dict_.values()))

disc = SimpleDiscretizer(n_bins=4, strategy='quantile')
disc.fit(X, feature_placeholders)
X_disc = disc.transform(X)
```
